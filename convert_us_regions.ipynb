```python
lines_lst = []
with open('f:/datasets/us_regions.txt', mode='r') as src:
    next(src)
    for line in src:
        line = line.strip()
        line = line.split()
        if len(line) > 1 and len(line) < 4:
            lines_lst.append(line)
        elif len(line) > 3 and len(line) < 5:
            # print(line)
            line_0 = line[0]
            line_1 = line[1] + ' ' + line[2]
            line_2 = line[3]
            line = "['{0}', '{1}', '{2}']".format(line_0, line_1, line_2)
            # print(line)
            lines_lst.append(line)
        elif len(line) >=5:
            print(line)
            line_0 = line[0]
            line_1 = line[1] + ' ' + line[2] + ' ' + line[3]
            line_2 = line[4]
            line = "['{0}', '{1}', '{2}']".format(line_0, line_1, line_2)
            lines_lst.append(line)
            
                                
        
        
```

```python
for line in lines_lst:
    print(line)
```

```python
with open("f:datasets/us_regions_2.csv", mode="w") as f:
    for line in lines_lst:
        line = str(line).replace('[', '').replace(']', '')
        print(line)
        print(line, file=f)
```
